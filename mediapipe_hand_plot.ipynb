```python
import cv2
import mediapipe as mp
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

import plotly.graph_objects as go

#import chart_studio

#chart_studio.tools.set_credentials_file(username='', api_key='')

mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_hands = mp.solutions.hands

# For static images:
IMAGE_FILES = ['/content/drive/MyDrive/Colab Notebooks/hand1.jpg']
with mp_hands.Hands(
    static_image_mode=True,
    max_num_hands=1,
    min_detection_confidence=0.5) as hands:
  for idx, file in enumerate(IMAGE_FILES):
    # Read an image, flip it around y-axis for correct handedness output (see
    # above).
    image = cv2.flip(cv2.imread(file), 1)
    # Convert the BGR image to RGB before processing.
    results = hands.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    if results.multi_hand_landmarks:
      for hand_landmarks in results.multi_hand_landmarks:
        #print(hand_landmarks.landmark[0])
        fig = plt.figure(figsize=(10,10))

        plotly_fig = go.Figure()

        ax = fig.add_subplot(111, projection='3d')

        for i in range(0,21):
          #hand_landmarks.landmark[i].x = hand_landmarks.landmark[i].x * -1
          #hand_landmarks.landmark[i].y = hand_landmarks.landmark[i].y * -1
          #hand_landmarks.landmark[i].z = hand_landmarks.landmark[i].z * -1
          ax.scatter(hand_landmarks.landmark[i].x,hand_landmarks.landmark[i].y,hand_landmarks.landmark[i].z)
          plotly_fig.add_trace(go.Scatter3d(
              x=[hand_landmarks.landmark[i].x], 
              y=[hand_landmarks.landmark[i].y], 
              z=[hand_landmarks.landmark[i].z]))
          
        for i in range(0,4):
          ax.plot(
              np.linspace(hand_landmarks.landmark[i].x,hand_landmarks.landmark[i+1].x),
              np.linspace(hand_landmarks.landmark[i].y,hand_landmarks.landmark[i+1].y) ,
              np.linspace(hand_landmarks.landmark[i].z,hand_landmarks.landmark[i+1].z)
              )
          plotly_fig.add_trace(go.Scatter3d(
              x=[hand_landmarks.landmark[i].x,hand_landmarks.landmark[i+1].x], 
              y=[hand_landmarks.landmark[i].y,hand_landmarks.landmark[i+1].y], 
              z=[hand_landmarks.landmark[i].z,hand_landmarks.landmark[i+1].z], 
              mode="lines"
              ))
        
        for i in range(5,8):
          ax.plot(
              np.linspace(hand_landmarks.landmark[i].x,hand_landmarks.landmark[i+1].x),
              np.linspace(hand_landmarks.landmark[i].y,hand_landmarks.landmark[i+1].y) ,
              np.linspace(hand_landmarks.landmark[i].z,hand_landmarks.landmark[i+1].z)
              )
          plotly_fig.add_trace(go.Scatter3d(
              x=[hand_landmarks.landmark[i].x,hand_landmarks.landmark[i+1].x], 
              y=[hand_landmarks.landmark[i].y,hand_landmarks.landmark[i+1].y], 
              z=[hand_landmarks.landmark[i].z,hand_landmarks.landmark[i+1].z], 
              mode="lines"
              ))
          
        for i in range(9,12):
          ax.plot(
              np.linspace(hand_landmarks.landmark[i].x,hand_landmarks.landmark[i+1].x),
              np.linspace(hand_landmarks.landmark[i].y,hand_landmarks.landmark[i+1].y) ,
              np.linspace(hand_landmarks.landmark[i].z,hand_landmarks.landmark[i+1].z)
              )
          plotly_fig.add_trace(go.Scatter3d(
              x=[hand_landmarks.landmark[i].x,hand_landmarks.landmark[i+1].x], 
              y=[hand_landmarks.landmark[i].y,hand_landmarks.landmark[i+1].y], 
              z=[hand_landmarks.landmark[i].z,hand_landmarks.landmark[i+1].z], 
              mode="lines"
              ))
          
        for i in range(13,16):
          ax.plot(
              np.linspace(hand_landmarks.landmark[i].x,hand_landmarks.landmark[i+1].x),
              np.linspace(hand_landmarks.landmark[i].y,hand_landmarks.landmark[i+1].y) ,
              np.linspace(hand_landmarks.landmark[i].z,hand_landmarks.landmark[i+1].z)
              )
          plotly_fig.add_trace(go.Scatter3d(
              x=[hand_landmarks.landmark[i].x,hand_landmarks.landmark[i+1].x], 
              y=[hand_landmarks.landmark[i].y,hand_landmarks.landmark[i+1].y], 
              z=[hand_landmarks.landmark[i].z,hand_landmarks.landmark[i+1].z], 
              mode="lines"
              ))
          
        for i in range(17,20):
          ax.plot(
              np.linspace(hand_landmarks.landmark[i].x,hand_landmarks.landmark[i+1].x),
              np.linspace(hand_landmarks.landmark[i].y,hand_landmarks.landmark[i+1].y) ,
              np.linspace(hand_landmarks.landmark[i].z,hand_landmarks.landmark[i+1].z)
              )
          plotly_fig.add_trace(go.Scatter3d(
              x=[hand_landmarks.landmark[i].x,hand_landmarks.landmark[i+1].x], 
              y=[hand_landmarks.landmark[i].y,hand_landmarks.landmark[i+1].y], 
              z=[hand_landmarks.landmark[i].z,hand_landmarks.landmark[i+1].z], 
              mode="lines"
              ))

        ax.plot(
              np.linspace(hand_landmarks.landmark[0].x,hand_landmarks.landmark[5].x),
              np.linspace(hand_landmarks.landmark[0].y,hand_landmarks.landmark[5].y) ,
              np.linspace(hand_landmarks.landmark[0].z,hand_landmarks.landmark[5].z)
              )

        ax.plot(
              np.linspace(hand_landmarks.landmark[0].x,hand_landmarks.landmark[17].x),
              np.linspace(hand_landmarks.landmark[0].y,hand_landmarks.landmark[17].y) ,
              np.linspace(hand_landmarks.landmark[0].z,hand_landmarks.landmark[17].z)
              )
        
        ax.plot(
              np.linspace(hand_landmarks.landmark[5].x,hand_landmarks.landmark[9].x),
              np.linspace(hand_landmarks.landmark[5].y,hand_landmarks.landmark[9].y) ,
              np.linspace(hand_landmarks.landmark[5].z,hand_landmarks.landmark[9].z)
              )
        
        ax.plot(
              np.linspace(hand_landmarks.landmark[9].x,hand_landmarks.landmark[13].x),
              np.linspace(hand_landmarks.landmark[9].y,hand_landmarks.landmark[13].y) ,
              np.linspace(hand_landmarks.landmark[9].z,hand_landmarks.landmark[13].z)
              )
        
        ax.plot(
              np.linspace(hand_landmarks.landmark[13].x,hand_landmarks.landmark[17].x),
              np.linspace(hand_landmarks.landmark[13].y,hand_landmarks.landmark[17].y) ,
              np.linspace(hand_landmarks.landmark[13].z,hand_landmarks.landmark[17].z)
              )
        
        ########

        plotly_fig.add_trace(go.Scatter3d(
              x=[hand_landmarks.landmark[0].x,hand_landmarks.landmark[5].x], 
              y=[hand_landmarks.landmark[0].y,hand_landmarks.landmark[5].y], 
              z=[hand_landmarks.landmark[0].z,hand_landmarks.landmark[5].z], 
              mode="lines"
              ))
        
        plotly_fig.add_trace(go.Scatter3d(
              x=[hand_landmarks.landmark[0].x,hand_landmarks.landmark[17].x], 
              y=[hand_landmarks.landmark[0].y,hand_landmarks.landmark[17].y], 
              z=[hand_landmarks.landmark[0].z,hand_landmarks.landmark[17].z], 
              mode="lines"
              ))
        
        plotly_fig.add_trace(go.Scatter3d(
              x=[hand_landmarks.landmark[5].x,hand_landmarks.landmark[9].x], 
              y=[hand_landmarks.landmark[5].y,hand_landmarks.landmark[9].y], 
              z=[hand_landmarks.landmark[5].z,hand_landmarks.landmark[9].z], 
              mode="lines"
              ))
        
        plotly_fig.add_trace(go.Scatter3d(
              x=[hand_landmarks.landmark[9].x,hand_landmarks.landmark[13].x], 
              y=[hand_landmarks.landmark[9].y,hand_landmarks.landmark[13].y], 
              z=[hand_landmarks.landmark[9].z,hand_landmarks.landmark[13].z], 
              mode="lines"
              ))
        
        plotly_fig.add_trace(go.Scatter3d(
              x=[hand_landmarks.landmark[13].x,hand_landmarks.landmark[17].x], 
              y=[hand_landmarks.landmark[13].y,hand_landmarks.landmark[17].y], 
              z=[hand_landmarks.landmark[13].z,hand_landmarks.landmark[17].z], 
              mode="lines"
              ))

        plt.show()
        
        plotly_fig.show()
        
    # Draw hand world landmarks.
    if not results.multi_hand_world_landmarks:
      continue
    for hand_world_landmarks in results.multi_hand_world_landmarks:
      mp_drawing.plot_landmarks(
        hand_world_landmarks, mp_hands.HAND_CONNECTIONS, azimuth=5)
```
